Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans: ArithmeticError is simply an error that occurs during numeric calculations.

```python
# ZeroDivision Error is an arithmetic error which we have in a division with denominator as zero
num1 = 1
num2 = 0
try:
    num3 = num1 / num2
except ZeroDivisionError as ex:
    print("Error: ", ex)
```

```python
# OverflowError is an arithmetic error which occurs when the acceptable range is overflowed
j = 5.0
try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("<----Overflow error---->")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans) The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.
LookupError exception class can handle both IndexError and KeyError exception classes.

```python
# For Index error
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```

```python
# For Key error

my_dict = {'name': 'Vishnu Prasad',
            'age': 24,
            'language': 'Python'}

user_input = "framework"

try:
    print(f'{user_input} is {my_dict[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```
