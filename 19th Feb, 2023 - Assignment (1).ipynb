Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans) 
Flask is a lightweight web framework written in Python. It is designed to make it easy to build web applications quickly and with minimum overhead. Flask is often referred to as a "micro" framework because it does not require particular tools or libraries to function. 

Some of the advantages of using Flask include:

1. Lightweight and flexible: Flask is a lightweight framework that is easy to use and flexible. It does not impose any restrictions on the way you structure your application, allowing you to build applications the way you want.

2. Modular design: Flask has a modular design, allowing developers to use only the components they need. This makes it easy to add new features to an existing application or to remove features that are no longer needed.

3. Easy to learn: Flask is easy to learn, even for developers who are new to web development. The framework provides a simple and intuitive API, making it easy to get started and build applications quickly.

4. Large and supportive community: Flask has a large and supportive community of developers who contribute to the framework and provide support to other developers. This means that you can find plenty of resources, tutorials, and documentation online to help you build your application.

5. Extensible: Flask can be easily extended with third-party libraries to add additional functionality to your application. There are many extensions available for Flask, including libraries for database integration, authentication, and more.

Q3. What is App routing in Flask? Why do we use app routes?

Ans) 
In Flask, app routing refers to the process of mapping URLs to functions that handle them. In other words, when a user visits a particular URL, Flask uses the app routing system to determine which function should be called to handle the request.

Flask app routing is typically done using a decorator called @app.route(). This decorator specifies the URL route that should be handled by the decorated function. For example, if you want to handle requests to the URL /hello, you would use the @app.route('/hello') decorator before defining the function that handles that URL.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans) 
In Flask, the url_for() function is used for URL building. This function generates a URL for the specified endpoint, which is defined by the name of a function in the application.

Example of how to use the url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'This is the contact page'

@app.route('/user/<username>')
def profile(username):
    return 'Hello, {}!'.format(username)

with app.test_request_context():
    print(url_for('home'))
    print(url_for('about'))
    print(url_for('contact'))
    print(url_for('profile', username='Vishnu'))
    
```

Output: 

/

/about

/contact

/user/Vishnu

the url_for() function generates URLs for each of the routes based on the function names we pass as parameters.
